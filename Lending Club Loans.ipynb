# DATA CLEANING

```python
new_clean_data = new.copy()
```

```python
new_clean_data.shape
```

```python
new_clean_data.info()
```

```python
# Here ['member id'] is in float 
# But the member id can't be in float 
# so i prefer to change it into integer
```

```python
new_clean_data['member_id'] = new_clean_data['member_id'].astype('int')
```

```python
#Here change int rate in to 1st float and then 2nd in percentage
```

```python
new_clean_data.info()
```

```python
new_clean_data['emp_length'].value_counts()
```

```python
new_clean_data['emp_length_year']=new_clean_data['emp_length']
```

```python
import re
```

```python
def nums(x):
    num = re.compile(r'[\d.]+').findall(x)
    return num[0]
```

```python
new_clean_data['int_rate_%']=new_clean_data['int_rate'].apply(nums)
```

```python
new_clean_data['int_rate_%']=pd.to_numeric(new_clean_data['int_rate_%'])
```

```python
new_clean_data['int_rate_%'].dtype
```

```python
new_clean_data.drop(columns=['int_rate'],axis=1, inplace=True)
```

```python
new_clean_data.drop(columns=['emp_length'], axis=1, inplace=True)
```

```python
new_clean_data['revol_util'].value_counts()
```

```python
new_clean_data['revol_util_%']=new_clean_data['revol_util'].apply(nums)
```

```python
new_clean_data['revol_util_%']=new_clean_data['revol_util_%'].astype('float64')
```

```python
new_clean_data.drop(columns=['revol_util'],axis=1, inplace=True)
```

```python
new_clean_data['emp_length_year']
```

```python
new_clean_data['total_pymnt'] # making the value rouundup
```

```python
new_clean_data['total_payment']=new_clean_data['total_pymnt'].apply(lambda x:np.round(x, decimals=2))
```
